## 1.Intervalo de Confianza para la Media Poblacional

Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes, se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm. Se construirá un intervalo de confianza del 95 % para la media poblacional.

```python
import numpy as np
from scipy.stats import norm

# 1. Datos proporcionados
n_sample = 50  # Tamaño de la muestra
mean_sample = 170  # Media de la muestra
std_dev_population = 10  # Desviación estándar poblacional
confidence_level = 0.95  # Nivel de confianza
alpha = 1 - confidence_level # Nivel de significancia

# 2. Identificar el estadístico de prueba y su distribución
# Dado que se conoce la desviación estándar poblacional y el tamaño de la muestra es grande (n > 30),
# utilizamos el estadístico Z y la distribución normal estándar.

# 3. Calcular el error estándar de la media
# Fórmula: SE = σ / sqrt(n)
standard_error = std_dev_population / np.sqrt(n_sample)
print(f"Error Estándar (SE): {standard_error:.4f}")

# 4. Determinar el valor crítico (Z) para el nivel de confianza
# Para un intervalo de confianza bilateral, el área en cada cola es alpha/2.
# Usamos la función ppf (percent point function) de la distribución normal para encontrar el valor Z.
critical_z_value = norm.ppf(1 - alpha/2)
print(f"Valor Crítico Z: {critical_z_value:.4f}")

# 5. Calcular el margen de error
# Fórmula: ME = Z * SE
margin_of_error = critical_z_value * standard_error
print(f"Margen de Error (ME): {margin_of_error:.4f}")

# 6. Construir el intervalo de confianza
# Fórmula: IC = media_muestra ± Margen_de_Error
confidence_interval_lower = mean_sample - margin_of_error
confidence_interval_upper = mean_sample + margin_of_error

print(f"\nIntervalo de Confianza del {confidence_level*100}%:")
print(f"[{confidence_interval_lower:.2f}, {confidence_interval_upper:.2f}]")
```

**Interpretación del Intervalo de Confianza:**

Con un 95% de confianza, la verdadera altura promedio de todos los estudiantes universitarios se encuentra entre approximately ${confidence_interval_lower:.2f}$ cm y ${confidence_interval_upper:.2f}$ cm. Esto significa que si tomáramos muchas muestras de 50 estudiantes y construyéramos un intervalo de confianza para cada una, el 95% de esos intervalos contendrían la verdadera media poblacional.

Una cadena de cafeterías desea estimar la temperatura promedio (°C) del café que sirve a sus clientes. Se tomó una muestra aleatoria de 12 tazas donde la media fue de 84°C, y la desviación estándar 5°C.
```python
import numpy as np
from scipy.stats import t

# 1. Datos proporcionados
n_sample_coffee = 12  # Tamaño de la muestra
mean_sample_coffee = 84  # Media de la muestra
std_dev_sample_coffee = 5  # Desviación estándar de la muestra
confidence_level_coffee = 0.95  # Nivel de confianza
alpha_coffee = 1 - confidence_level_coffee # Nivel de significancia
degrees_of_freedom_coffee = n_sample_coffee - 1 # Grados de libertad para la distribución t

# 2. Identificar el estadístico de prueba y su distribución
# Dado que se desconoce la desviación estándar poblacional y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico t de Student.

# 3. Calcular el error estándar de la media
# Fórmula: SE = s / sqrt(n)
standard_error_coffee = std_dev_sample_coffee / np.sqrt(n_sample_coffee)
print(f"Error Estándar (SE): {standard_error_coffee:.4f}")

# 4. Determinar el valor crítico (t) para el nivel de confianza y grados de libertad
# Para un intervalo de confianza bilateral, el área en cada cola es alpha/2.
# Usamos la función ppf (percent point function) de la distribución t para encontrar el valor t.
critical_t_value_coffee = t.ppf(1 - alpha_coffee/2, degrees_of_freedom_coffee)
print(f"Valor Crítico t: {critical_t_value_coffee:.4f}")

# 5. Calcular el margen de error
# Fórmula: ME = t * SE
margin_of_error_coffee = critical_t_value_coffee * standard_error_coffee
print(f"Margen de Error (ME): {margin_of_error_coffee:.4f}")

# 6. Construir el intervalo de confianza
# Fórmula: IC = media_muestra ± Margen_de_Error
confidence_interval_lower_coffee = mean_sample_coffee - margin_of_error_coffee
confidence_interval_upper_coffee = mean_sample_coffee + margin_of_error_coffee

print(f"\nIntervalo de Confianza del {confidence_level_coffee*100}%:")
print(f"[{confidence_interval_lower_coffee:.2f}, {confidence_interval_upper_coffee:.2f}]")
```

La biblioteca de una universidad necesita estimar el peso promedio (kg) de los libros de matemáticas. Tomó una muestra aleatoria de 8 libros los cuales tuvieron una media de 1.5 kg y una desviación estándar de 0.3 kg. Calcule el intervalo de confianza para el peso medio poblacional de los libros con un nivel de confianza del 95 %.

```python
import numpy as np
from scipy.stats import t

# 1. Datos proporcionados
n_sample_books = 8  # Tamaño de la muestra
mean_sample_books = 1.5  # Media de la muestra
std_dev_sample_books = 0.3  # Desviación estándar de la muestra
confidence_level_books = 0.95  # Nivel de confianza
alpha_books = 1 - confidence_level_books # Nivel de significancia
degrees_of_freedom_books = n_sample_books - 1 # Grados de libertad para la distribución t

# 2. Identificar el estadístico de prueba y su distribución
# Dado que se desconoce la desviación estándar poblacional y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico t de Student.

# 3. Calcular el error estándar de la media
# Fórmula: SE = s / sqrt(n)
standard_error_books = std_dev_sample_books / np.sqrt(n_sample_books)
print(f"Error Estándar (SE): {standard_error_books:.4f}")

# 4. Determinar el valor crítico (t) para el nivel de confianza y grados de libertad
# Para un intervalo de confianza bilateral, el área en cada cola es alpha/2.
# Usamos la función ppf (percent point function) de la distribución t para encontrar el valor t.
critical_t_value_books = t.ppf(1 - alpha_books/2, degrees_of_freedom_books)
print(f"Valor Crítico t: {critical_t_value_books:.4f}")

# 5. Calcular el margen de error
# Fórmula: ME = t * SE
margin_of_error_books = critical_t_value_books * standard_error_books
print(f"Margen de Error (ME): {margin_of_error_books:.4f}")

# 6. Construir el intervalo de confianza
# Fórmula: IC = media_muestra ± Margen_de_Error
confidence_interval_lower_books = mean_sample_books - margin_of_error_books
confidence_interval_upper_books = mean_sample_books + margin_of_error_books

print(f"\nIntervalo de Confianza del {confidence_level_books*100}%:")
print(f"[{confidence_interval_lower_books:.3f}, {confidence_interval_upper_books:.3f}]")
```

**Interpretación del Intervalo de Confianza:**

Con un 95% de confianza, el verdadero peso promedio de los libros de matemáticas en la biblioteca se encuentra entre approximately ${confidence_interval_lower_books:.3f}$ kg y ${confidence_interval_upper_books:.3f}$ kg. Esto significa que si tomáramos muchas muestras de 8 libros de matemáticas y construyéramos un intervalo de confianza para cada una, el 95% de esos intervalos contendrán la verdadera media poblacional.

## 21. Prueba de Hipótesis para la Media Poblacional (Peso de Cereal)

Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia α = 0,05.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 500 g (El peso promedio es 500 g)
# Ha: μ < 500 g (El peso promedio es menor a 500 g - sospecha de subllenado)
mu_population = 500  # Media poblacional bajo H0
alpha = 0.05  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: μ = {mu_population} g")
print(f"   Ha: μ < {mu_population} g (Prueba de cola izquierda)")

# 2. Datos de la muestra y desviación estándar poblacional
n_sample = 40  # Tamaño de la muestra
mean_sample = 495  # Media de la muestra
std_dev_population = 15  # Desviación estándar poblacional

# Identificar el estadístico de prueba y su distribución
# Dado que se conoce la desviación estándar poblacional y el tamaño de la muestra es grande (n > 30),
# utilizamos el estadístico Z y la distribución normal estándar.

# 3. Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestra - μ_poblacional) / (σ_poblacional / sqrt(n))
standard_error = std_dev_population / np.sqrt(n_sample)
z_statistic = (mean_sample - mu_population) / standard_error
print(f"\n2. Estadístico de prueba Z: {z_statistic:.4f}")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola izquierda con nivel de significancia α, el valor crítico Z es el cuantil α.
critical_z_value = norm.ppf(alpha)
print(f"3. Valor Crítico Z para α = {alpha}: {critical_z_value:.4f}")
print(f"   Región de Rechazo: Z < {critical_z_value:.4f}")

# Calcular el valor p
# Para una prueba de cola izquierda, el valor p es la probabilidad de obtener un estadístico Z tan extremo o más extremo (menor) que el calculado,
# asumiendo que H0 es verdadera.
p_value = norm.cdf(z_statistic)
print(f"   Valor p: {p_value:.4f}")


# 6. Graficar la distribución, región de rechazo y estadístico
x = np.linspace(-4, 4, 200)
y = norm.pdf(x, 0, 1)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(-4, critical_z_value, 100)
y_reject = norm.pdf(x_reject, 0, 1)
ax.fill_between(x_reject, y_reject, color='blue', alpha=0.5, label=f'Región de Rechazo (α={alpha})')

# Dibujar el valor crítico
ax.axvline(critical_z_value, color='blue', linestyle='--', label=f'Valor Crítico Z = {critical_z_value:.4f}')

# Dibujar el estadístico calculado
ax.axvline(z_statistic, color='green', linestyle='-', label=f'Estadístico Z = {z_statistic:.4f}')

ax.set_xlabel('Valor Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Región de Rechazo y Estadístico de Prueba')
ax.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n7. Decisión Estadística:")
if z_statistic < critical_z_value:
    print(f"   Como el estadístico Z ({z_statistic:.4f}) es menor que el valor crítico Z ({critical_z_value:.4f}),")
    print("   se rechaza la hipótesis nula (H0).")
    decision = "Rechazar H0"
else:
    print(f"   Como el estadístico Z ({z_statistic:.4f}) no es menor que el valor crítico Z ({critical_z_value:.4f}),")
    print("   no se rechaza la hipótesis nula (H0).")
    decision = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha}, hay suficiente evidencia para concluir que el peso promedio de los paquetes de cereal es menor a 500 g.")
    print(f"El valor p ({p_value:.4f}) es menor que α ({alpha}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha}, no hay suficiente evidencia para concluir que el peso promedio de los paquetes de cereal es menor a 500 g.")
    print(f"El valor p ({p_value:.4f}) es mayor que α ({alpha}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value:.4f} representa la probabilidad de observar una media muestral de 495 g (o menos), asumiendo que el verdadero peso promedio poblacional es 500 g.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

## 23. Prueba de Hipótesis para la Media Poblacional (Promedio de Clase en Examen Nacional)

Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 75 puntos (El promedio de la clase es igual al promedio nacional)
# Ha: μ ≠ 75 puntos (El promedio de la clase es diferente al promedio nacional - prueba de dos colas)
mu_population_class_exam = 75  # Media poblacional bajo H0
alpha_class_exam = 0.05  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: μ = {mu_population_class_exam} puntos")
print(f"   Ha: μ ≠ {mu_population_class_exam} puntos (Prueba de dos colas)")

# 2. Datos de la muestra y desviación estándar poblacional
n_sample_class_exam = 36  # Tamaño de la muestra
mean_sample_class_exam = 78  # Media de la muestra
std_dev_population_class_exam = 12  # Desviación estándar poblacional

# Identificar el estadístico de prueba y su distribución
# Dado que se conoce la desviación estándar poblacional y el tamaño de la muestra es grande (n > 30),
# utilizamos el estadístico Z y la distribución normal estándar.

# 3. Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestra - μ_poblacional) / (σ_poblacional / sqrt(n))
standard_error_class_exam = std_dev_population_class_exam / np.sqrt(n_sample_class_exam)
z_statistic_class_exam = (mean_sample_class_exam - mu_population_class_exam) / standard_error_class_exam
print(f"\n2. Estadístico de prueba Z: {z_statistic_class_exam:.4f}")

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba de dos colas con nivel de significancia α, el área en cada cola es α/2.
# Hay dos valores críticos, uno positivo y uno negativo.
critical_z_value_right_class_exam = norm.ppf(1 - alpha_class_exam/2)
critical_z_value_left_class_exam = norm.ppf(alpha_class_exam/2)
print(f"3. Valores Críticos Z para α = {alpha_class_exam}: {critical_z_value_left_class_exam:.4f} y {critical_z_value_right_class_exam:.4f}")
print(f"   Región de Rechazo: Z < {critical_z_value_left_class_exam:.4f} o Z > {critical_z_value_right_class_exam:.4f}")

# Calcular el valor p
# Para una prueba de dos colas, el valor p es el doble de la probabilidad de obtener un estadístico Z tan extremo o más extremo que el calculado (en cualquier dirección),
# asumiendo que H0 es verdadera.
p_value_class_exam = 2 * (1 - norm.cdf(abs(z_statistic_class_exam))) # Usamos abs() para considerar ambas colas
print(f"   Valor p: {p_value_class_exam:.4f}")

# 6. Graficar la distribución, regiones de rechazo y estadístico
x = np.linspace(-4, 4, 200)
y = norm.pdf(x, 0, 1)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (dos colas)
x_reject_left = np.linspace(-4, critical_z_value_left_class_exam, 100)
y_reject_left = norm.pdf(x_reject_left, 0, 1)
ax.fill_between(x_reject_left, y_reject_left, color='blue', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_class_exam/2})')

x_reject_right = np.linspace(critical_z_value_right_class_exam, 4, 100)
y_reject_right = norm.pdf(x_reject_right, 0, 1)
ax.fill_between(x_reject_right, y_reject_right, color='blue', alpha=0.5) # No duplicar la etiqueta

# Dibujar los valores críticos
ax.axvline(critical_z_value_left_class_exam, color='blue', linestyle='--', label=f'Valores Críticos Z = {critical_z_value_left_class_exam:.4f}, {critical_z_value_right_class_exam:.4f}')
ax.axvline(critical_z_value_right_class_exam, color='blue', linestyle='--') # No duplicar la etiqueta

# Dibujar el estadístico calculado
ax.axvline(z_statistic_class_exam, color='green', linestyle='-', label=f'Estadístico Z = {z_statistic_class_exam:.4f}')

ax.set_xlabel('Valor Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba (Prueba de Dos Colas)')
ax.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n7. Decisión Estadística:")
if z_statistic_class_exam < critical_z_value_left_class_exam or z_statistic_class_exam > critical_z_value_right_class_exam:
    print(f"   Como el estadístico Z ({z_statistic_class_exam:.4f}) cae en la región de rechazo,")
    print("   se rechaza la hipótesis nula (H0).")
    decision_class_exam = "Rechazar H0"
else:
    print(f"   Como el estadístico Z ({z_statistic_class_exam:.4f}) no cae en la región de rechazo,")
    print("   no se rechaza la hipótesis nula (H0).")
    decision_class_exam = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision_class_exam == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha_class_exam}, hay suficiente evidencia para concluir que el promedio de la clase es diferente al promedio nacional de 75 puntos.")
    print(f"El valor p ({p_value_class_exam:.4f}) es menor que α ({alpha_class_exam}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha_class_exam}, no hay suficiente evidencia para concluir que el promedio de la clase es diferente al promedio nacional de 75 puntos.")
    print(f"El valor p ({p_value_class_exam:.4f}) es mayor que α ({alpha_class_exam}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value_class_exam:.4f} representa la probabilidad de observar una media muestral de 78 puntos (o más) o 72 puntos (o menos), asumiendo que el verdadero promedio poblacional es 75 puntos.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision_class_exam == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```

## 25. Prueba de Hipótesis para la Media Poblacional (Volumen de Llenado de Botellas)

Una máquina llena botellas de 500 mL. Una auditoría revela que, en 60 botellas, la media es 498 mL. Se sabe que σ = 5 mL. Con un nivel de significancia: α = 0,01.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 500 mL (El volumen promedio de llenado es 500 mL)
# Ha: μ ≠ 500 mL (El volumen promedio de llenado es diferente a 500 mL - prueba de dos colas)
mu_population_bottles = 500  # Media poblacional bajo H0
alpha_bottles_ht = 0.01  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: μ = {mu_population_bottles} mL")
print(f"   Ha: μ ≠ {mu_population_bottles} mL (Prueba de dos colas)")

# 2. Datos de la muestra y desviación estándar poblacional
n_sample_bottles_ht = 60  # Tamaño de la muestra
mean_sample_bottles_ht = 498  # Media de la muestra
std_dev_population_bottles_ht = 5  # Desviación estándar poblacional

# Identificar el estadístico de prueba y su distribución
# Dado que se conoce la desviación estándar poblacional y el tamaño de la muestra es grande (n > 30),
# utilizamos el estadístico Z y la distribución normal estándar.

# 3. Calcular el estadístico de prueba Z
# Fórmula: Z = (media_muestra - μ_poblacional) / (σ_poblacional / sqrt(n))
standard_error_bottles_ht = std_dev_population_bottles_ht / np.sqrt(n_sample_bottles_ht)
z_statistic_bottles_ht = (mean_sample_bottles_ht - mu_population_bottles) / standard_error_bottles_ht
print(f"\n2. Estadístico de prueba Z: {z_statistic_bottles_ht:.4f}")

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba de dos colas con nivel de significancia α, el área en cada cola es α/2.
# Hay dos valores críticos, uno positivo y uno negativo.
critical_z_value_right_bottles_ht = norm.ppf(1 - alpha_bottles_ht/2)
critical_z_value_left_bottles_ht = norm.ppf(alpha_bottles_ht/2)
print(f"3. Valores Críticos Z para α = {alpha_bottles_ht}: {critical_z_value_left_bottles_ht:.4f} y {critical_z_value_right_bottles_ht:.4f}")
print(f"   Región de Rechazo: Z < {critical_z_value_left_bottles_ht:.4f} o Z > {critical_z_value_right_bottles_ht:.4f}")

# Calcular el valor p
# Para una prueba de dos colas, el valor p es el doble de la probabilidad de obtener un estadístico Z tan extremo o más extremo que el calculado (en cualquier dirección),
# asumiendo que H0 es verdadera.
p_value_bottles_ht = 2 * (1 - norm.cdf(abs(z_statistic_bottles_ht))) # Usamos abs() para considerar ambas colas
print(f"   Valor p: {p_value_bottles_ht:.4f}")

# 6. Graficar la distribución, regiones de rechazo y estadístico
x = np.linspace(-4, 4, 200)
y = norm.pdf(x, 0, 1)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label='Distribución Normal Estándar')

# Rellenar la región de rechazo (dos colas)
x_reject_left = np.linspace(-4, critical_z_value_left_bottles_ht, 100)
y_reject_left = norm.pdf(x_reject_left, 0, 1)
ax.fill_between(x_reject_left, y_reject_left, color='blue', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_bottles_ht/2})')

x_reject_right = np.linspace(critical_z_value_right_bottles_ht, 4, 100)
y_reject_right = norm.pdf(x_reject_right, 0, 1)
ax.fill_between(x_reject_right, y_reject_right, color='blue', alpha=0.5) # No duplicar la etiqueta

# Dibujar los valores críticos
ax.axvline(critical_z_value_left_bottles_ht, color='blue', linestyle='--', label=f'Valores Críticos Z = {critical_z_value_left_bottles_ht:.4f}, {critical_z_value_right_bottles_ht:.4f}')
ax.axvline(critical_z_value_right_bottles_ht, color='blue', linestyle='--') # No duplicar la etiqueta

# Dibujar el estadístico calculado
ax.axvline(z_statistic_bottles_ht, color='green', linestyle='-', label=f'Estadístico Z = {z_statistic_bottles_ht:.4f}')

ax.set_xlabel('Valor Z')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Normal Estándar con Regiones de Rechazo y Estadístico de Prueba (Prueba de Dos Colas)')
ax.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n7. Decisión Estadística:")
if z_statistic_bottles_ht < critical_z_value_left_bottles_ht or z_statistic_bottles_ht > critical_z_value_right_bottles_ht:
    print(f"   Como el estadístico Z ({z_statistic_bottles_ht:.4f}) cae en la región de rechazo,")
    print("   se rechaza la hipótesis nula (H0).")
    decision_bottles_ht = "Rechazar H0"
else:
    print(f"   Como el estadístico Z ({z_statistic_bottles_ht:.4f}) no cae en la región de rechazo,")
    print("   no se rechaza la hipótesis nula (H0).")
    decision_bottles_ht = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision_bottles_ht == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha_bottles_ht}, hay suficiente evidencia para concluir que el volumen promedio de llenado de las botellas es diferente a 500 mL.")
    print(f"El valor p ({p_value_bottles_ht:.4f}) es menor que α ({alpha_bottles_ht}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha_bottles_ht}, no hay suficiente evidencia para concluir que el volumen promedio de llenado de las botellas es diferente a 500 mL.")
    print(f"El valor p ({p_value_bottles_ht:.4f}) es mayor que α ({alpha_bottles_ht}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value_bottles_ht:.4f} representa la probabilidad de observar una media muestral de 498 mL (o más extremo en cualquier dirección), asumiendo que el verdadero volumen promedio poblacional es 500 mL.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision_bottles_ht == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```

## 26. Prueba de Hipótesis para la Media Poblacional (Tiempo de Entrega a Domicilio)

Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: μ = 25 minutos (El tiempo promedio de entrega es 25 minutos)
# Ha: μ > 25 minutos (El tiempo promedio de entrega es mayor a 25 minutos - sospecha del cliente)
mu_population_delivery = 25  # Media poblacional bajo H0
alpha_delivery_ht = 0.05  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: μ = {mu_population_delivery} minutos")
print(f"   Ha: μ > {mu_population_delivery} minutos (Prueba de cola derecha)")

# 2. Datos de la muestra y desviación estándar muestral
n_sample_delivery_ht = 15  # Tamaño de la muestra
mean_sample_delivery_ht = 28  # Media de la muestra
std_dev_sample_delivery_ht = 4  # Desviación estándar de la muestra
degrees_of_freedom_delivery_ht = n_sample_delivery_ht - 1 # Grados de libertad para la distribución t

# Identificar el estadístico de prueba y su distribución
# Dado que se desconoce la desviación estándar poblacional y el tamaño de la muestra es pequeño (n < 30),
# utilizamos el estadístico t de Student.

# 3. Calcular el estadístico de prueba t
# Fórmula: t = (media_muestra - μ_poblacional) / (s_muestral / sqrt(n))
standard_error_delivery_ht = std_dev_sample_delivery_ht / np.sqrt(n_sample_delivery_ht)
t_statistic_delivery = (mean_sample_delivery_ht - mu_population_delivery) / standard_error_delivery_ht
print(f"\n2. Estadístico de prueba t: {t_statistic_delivery:.4f}")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola derecha con nivel de significancia α y grados de libertad, el valor crítico t es el cuantil 1-α.
critical_t_value_delivery_ht = t.ppf(1 - alpha_delivery_ht, degrees_of_freedom_delivery_ht)
print(f"3. Grados de Libertad: {degrees_of_freedom_delivery_ht}")
print(f"   Valor Crítico t para α = {alpha_delivery_ht}: {critical_t_value_delivery_ht:.4f}")
print(f"   Región de Rechazo: t > {critical_t_value_delivery_ht:.4f}")

# Calcular el valor p
# Para una prueba de cola derecha, el valor p es la probabilidad de obtener un estadístico t tan extremo o más extremo (mayor) que el calculado,
# asumiendo que H0 es verdadera.
p_value_delivery = 1 - t.cdf(t_statistic_delivery, degrees_of_freedom_delivery_ht)
print(f"   Valor p: {p_value_delivery:.4f}")

# 6. Graficar la distribución t, región de rechazo y estadístico
x = np.linspace(-4, 4, 200)
y = t.pdf(x, degrees_of_freedom_delivery_ht)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label=f'Distribución t (gl={degrees_of_freedom_delivery_ht})')

# Rellenar la región de rechazo (cola derecha)
x_reject = np.linspace(critical_t_value_delivery_ht, 4, 100)
y_reject = t.pdf(x_reject, degrees_of_freedom_delivery_ht)
ax.fill_between(x_reject, y_reject, color='blue', alpha=0.5, label=f'Región de Rechazo (α={alpha_delivery_ht})')

# Dibujar el valor crítico
ax.axvline(critical_t_value_delivery_ht, color='blue', linestyle='--', label=f'Valor Crítico t = {critical_t_value_delivery_ht:.4f}')

# Dibujar el estadístico calculado
ax.axvline(t_statistic_delivery, color='green', linestyle='-', label=f'Estadístico t = {t_statistic_delivery:.4f}')

ax.set_xlabel('Valor t')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución t con Región de Rechazo y Estadístico de Prueba (Prueba de Cola Derecha)')
ax.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("\n7. Decisión Estadística:")
if t_statistic_delivery > critical_t_value_delivery_ht:
    print(f"   Como el estadístico t ({t_statistic_delivery:.4f}) es mayor que el valor crítico t ({critical_t_value_delivery_ht:.4f}),")
    print("   se rechaza la hipótesis nula (H0).")
    decision_delivery = "Rechazar H0"
else:
    print(f"   Como el estadístico t ({t_statistic_delivery:.4f}) no es mayor que el valor crítico t ({critical_t_value_delivery_ht:.4f}),")
    print("   no se rechaza la hipótesis nula (H0).")
    decision_delivery = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision_delivery == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha_delivery_ht}, hay suficiente evidencia para concluir que el tiempo promedio de entrega a domicilio es mayor a {mu_population_delivery} minutos.")
    print(f"El valor p ({p_value_delivery:.4f}) es menor que α ({alpha_delivery_ht}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha_delivery_ht}, no hay suficiente evidencia para concluir que el tiempo promedio de entrega a domicilio es mayor a {mu_population_delivery} minutos.")
    print(f"El valor p ({p_value_delivery:.4f}) es mayor que α ({alpha_delivery_ht}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value_delivery:.4f} representa la probabilidad de observar una media muestral de 28 minutos (o más), asumiendo que el verdadero tiempo promedio poblacional es 25 minutos.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision_delivery == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```

## 31. Prueba de Hipótesis para la Varianza Poblacional

Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de σ² = 2,5 minutos². Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral de s² = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 2.5 (La varianza poblacional es 2.5)
# Ha: σ² ≠ 2.5 (La varianza poblacional difiere de 2.5 - prueba de dos colas)
sigma2_population = 2.5  # Varianza poblacional bajo H0
alpha = 0.05  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: σ² = {sigma2_population} minutos²")
print(f"   Ha: σ² ≠ {sigma2_population} minutos² (Prueba de dos colas)")

# 2. Datos de la muestra
n_sample = 16  # Tamaño de la muestra
s2_sample = 4  # Varianza de la muestra
degrees_of_freedom = n_sample - 1 # Grados de libertad

# 3. Calcular el estadístico de prueba Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi2_statistic = ((n_sample - 1) * s2_sample) / sigma2_population
print(f"\n2. Estadístico de prueba χ²: {chi2_statistic:.4f}")

# 4. Determinar los valores críticos y la región de rechazo
# Para una prueba de dos colas con nivel de significancia α y grados de libertad,
# hay dos valores críticos: α/2 en la cola inferior y 1-α/2 en la cola superior.
critical_chi2_lower = chi2.ppf(alpha/2, degrees_of_freedom)
critical_chi2_upper = chi2.ppf(1 - alpha/2, degrees_of_freedom)

print(f"3. Grados de Libertad: {degrees_of_freedom}")
print(f"   Valores Críticos χ² para α = {alpha}: {critical_chi2_lower:.4f} (inferior) y {critical_chi2_upper:.4f} (superior)")
print(f"   Región de Rechazo: χ² < {critical_chi2_lower:.4f} o χ² > {critical_chi2_upper:.4f}")

# Calcular el valor p
# Para una prueba de dos colas, el valor p es el doble del área en la cola más extrema.
# Si el estadístico está en la cola izquierda, p = 2 * P(χ² < estadístico).
# Si el estadístico está en la cola derecha, p = 2 * P(χ² > estadístico).
if chi2_statistic < degrees_of_freedom: # Aproximado: si está a la izquierda del centro de la distribución t
    p_value = 2 * chi2.cdf(chi2_statistic, degrees_of_freedom)
else:
    p_value = 2 * (1 - chi2.cdf(chi2_statistic, degrees_of_freedom))

print(f"   Valor p: {p_value:.4f}")


# 6. Graficar la distribución Chi-cuadrado, regiones de rechazo y estadístico
x = np.linspace(0, chi2.ppf(0.999, degrees_of_freedom), 200) # Extender el rango para ver la cola derecha si es necesario
y = chi2.pdf(x, degrees_of_freedom)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label=f'Distribución χ² (gl={degrees_of_freedom})')

# Rellenar las regiones de rechazo (dos colas)
x_reject_lower = np.linspace(0, critical_chi2_lower, 100)
y_reject_lower = chi2.pdf(x_reject_lower, degrees_of_freedom)
ax.fill_between(x_reject_lower, y_reject_lower, color='blue', alpha=0.5, label=f'Región de Rechazo (α/2={alpha/2})')

x_reject_upper = np.linspace(critical_chi2_upper, x[-1], 100)
y_reject_upper = chi2.pdf(x_reject_upper, degrees_of_freedom)
ax.fill_between(x_reject_upper, y_reject_upper, color='blue', alpha=0.5) # No duplicar la etiqueta

# Dibujar los valores críticos
ax.axvline(critical_chi2_lower, color='blue', linestyle='--', label=f'Valores Críticos χ² = {critical_chi2_lower:.4f}, {critical_chi2_upper:.4f}')
ax.axvline(critical_chi2_upper, color='blue', linestyle='--') # No duplicar la etiqueta

# Dibujar el estadístico calculado
ax.axvline(chi2_statistic, color='green', linestyle='-', label=f'Estadístico χ² = {chi2_statistic:.4f}')

ax.set_xlabel('Valor χ²')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución χ² con Regiones de Rechazo y Estadístico de Prueba (Prueba de Dos Colas)')
ax.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n4. Decisión Estadística:")
if chi2_statistic < critical_chi2_lower or chi2_statistic > critical_chi2_upper:
    print(f"   Como el estadístico χ² ({chi2_statistic:.4f}) cae en la región de rechazo,")
    print("   se rechaza la hipótesis nula (H0).")
    decision = "Rechazar H0"
else:
    print(f"   Como el estadístico χ² ({chi2_statistic:.4f}) no cae en la región de rechazo,")
    print("   no se rechaza la hipótesis nula (H0).")
    decision = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha}, hay suficiente evidencia para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {sigma2_population} minutos².")
    print(f"El valor p ({p_value:.4f}) es menor que α ({alpha}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha}, no hay suficiente evidencia para concluir que la varianza poblacional del tiempo de duración de las llamadas es diferente a {sigma2_population} minutos².")
    print(f"El valor p ({p_value:.4f}) es mayor que α ({alpha}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value:.4f} representa la probabilidad de observar una varianza muestral de {s2_sample} (o más extrema en cualquier dirección), asumiendo que la verdadera varianza poblacional es {sigma2_population}.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```

## 32. Prueba de Hipótesis para la Varianza Poblacional

Una máquina embotelladora debe mantener la varianza del llenado en σ² = 1,2 ml². Se toma una muestra de n = 20 botellas y se encuentra s² = 2,0. ¿Existe evidencia al 1 % de que la varianza es mayor?

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 1.2 (La varianza poblacional es 1.2)
# Ha: σ² > 1.2 (La varianza poblacional es mayor a 1.2 - prueba de cola derecha)
sigma2_population_bottles = 1.2  # Varianza poblacional bajo H0
alpha_bottles_var = 0.01  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: σ² = {sigma2_population_bottles} ml²")
print(f"   Ha: σ² > {sigma2_population_bottles} ml² (Prueba de cola derecha)")

# 2. Datos de la muestra
n_sample_bottles_var = 20  # Tamaño de la muestra
s2_sample_bottles_var = 2.0  # Varianza de la muestra
degrees_of_freedom_bottles_var = n_sample_bottles_var - 1 # Grados de libertad

# 3. Calcular el estadístico de prueba Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi2_statistic_bottles_var = ((n_sample_bottles_var - 1) * s2_sample_bottles_var) / sigma2_population_bottles
print(f"\n2. Estadístico de prueba χ²: {chi2_statistic_bottles_var:.4f}")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola derecha con nivel de significancia α y grados de libertad,
# el valor crítico χ² es el cuantil 1-α.
critical_chi2_bottles_var = chi2.ppf(1 - alpha_bottles_var, degrees_of_freedom_bottles_var)

print(f"3. Grados de Libertad: {degrees_of_freedom_bottles_var}")
print(f"   Valor Crítico χ² para α = {alpha_bottles_var}: {critical_chi2_bottles_var:.4f}")
print(f"   Región de Rechazo: χ² > {critical_chi2_bottles_var:.4f}")

# Calcular el valor p
# Para una prueba de cola derecha, el valor p es la probabilidad de obtener un estadístico χ² tan extremo o más extremo (mayor) que el calculado,
# asumiendo que H0 es verdadera.
p_value_bottles_var = 1 - chi2.cdf(chi2_statistic_bottles_var, degrees_of_freedom_bottles_var)

print(f"   Valor p: {p_value_bottles_var:.4f}")

# 6. Graficar la distribución Chi-cuadrado, región de rechazo y estadístico
x = np.linspace(0, chi2.ppf(0.999, degrees_of_freedom_bottles_var), 200) # Extender el rango para ver la cola derecha si es necesario
y = chi2.pdf(x, degrees_of_freedom_bottles_var)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label=f'Distribución χ² (gl={degrees_of_freedom_bottles_var})')

# Rellenar la región de rechazo (cola derecha)
x_reject = np.linspace(critical_chi2_bottles_var, x[-1], 100)
y_reject = chi2.pdf(x_reject, degrees_of_freedom_bottles_var)
ax.fill_between(x_reject, y_reject, color='blue', alpha=0.5, label=f'Región de Rechazo (α={alpha_bottles_var})')

# Dibujar el valor crítico
ax.axvline(critical_chi2_bottles_var, color='blue', linestyle='--', label=f'Valor Crítico χ² = {critical_chi2_bottles_var:.4f}')

# Dibujar el estadístico calculado
ax.axvline(chi2_statistic_bottles_var, color='green', linestyle='-', label=f'Estadístico χ² = {chi2_statistic_bottles_var:.4f}')

ax.set_xlabel('Valor χ²')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución χ² con Región de Rechazo y Estadístico de Prueba (Prueba de Cola Derecha)')
ax.legend()
plt.grid(True)
plt.show()

# 7. Tomar una decisión estadística y concluir
print("\n4. Decisión Estadística:")
if chi2_statistic_bottles_var > critical_chi2_bottles_var:
    print(f"   Como el estadístico χ² ({chi2_statistic_bottles_var:.4f}) es mayor que el valor crítico χ² ({critical_chi2_bottles_var:.4f}),")
    print("   se rechaza la hipótesis nula (H0).")
    decision_bottles_var = "Rechazar H0"
else:
    print(f"   Como el estadístico χ² ({chi2_statistic_bottles_var:.4f}) no es mayor que el valor crítico χ² ({critical_chi2_bottles_var:.4f}),")
    print("   no se rechaza la hipótesis nula (H0).")
    decision_bottles_var = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision_bottles_var == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha_bottles_var}, hay suficiente evidencia para concluir que la varianza poblacional del llenado es mayor a {sigma2_population_bottles} ml².")
    print(f"El valor p ({p_value_bottles_var:.4f}) es menor que α ({alpha_bottles_var}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha_bottles_var}, no hay suficiente evidencia para concluir que la varianza poblacional del llenado es mayor a {sigma2_population_bottles} ml².")
    print(f"El valor p ({p_value_bottles_var:.4f}) es mayor que α ({alpha_bottles_var}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value_bottles_var:.4f} representa la probabilidad de observar una varianza muestral de {s2_sample_bottles_var} (o mayor), asumiendo que la verdadera varianza poblacional es {sigma2_population_bottles}.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision_bottles_var == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```

## 33. Prueba de Hipótesis para la Varianza Poblacional

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# 1. Plantear las hipótesis nula y alternativa
# H0: σ² = 0.5 (La varianza poblacional es 0.5 - o mayor, pero la prueba es de cola izquierda)
# Ha: σ² < 0.5 (La varianza poblacional es menor a 0.5 - la afirmación a probar)
sigma2_population_thickness = 0.5  # Varianza poblacional bajo H0
alpha_thickness_var = 0.05  # Nivel de significancia

print("1. Hipótesis:")
print(f"   H0: σ² >= {sigma2_population_thickness} mm²") # En realidad H0 es sigma2 >= 0.5, pero para el cálculo se usa el límite inferior
print(f"   Ha: σ² < {sigma2_population_thickness} mm² (Prueba de cola izquierda)")

# 2. Datos de la muestra
n_sample_thickness_var = 12  # Tamaño de la muestra
s2_sample_thickness_var = 0.3  # Varianza de la muestra
degrees_of_freedom_thickness_var = n_sample_thickness_var - 1 # Grados de libertad

# 3. Calcular el estadístico de prueba Chi-cuadrado
# Fórmula: χ² = ((n - 1) * s²) / σ²
chi2_statistic_thickness_var = ((n_sample_thickness_var - 1) * s2_sample_thickness_var) / sigma2_population_thickness
print(f"\n2. Estadístico de prueba χ²: {chi2_statistic_thickness_var:.4f}")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola izquierda con nivel de significancia α y grados de libertad,
# el valor crítico χ² es el cuantil α.
critical_chi2_thickness_var = chi2.ppf(alpha_thickness_var, degrees_of_freedom_thickness_var)

print(f"3. Grados de Libertad: {degrees_of_freedom_thickness_var}")
print(f"   Valor Crítico χ² para α = {alpha_thickness_var}: {critical_chi2_thickness_var:.4f}")
print(f"   Región de Rechazo: χ² < {critical_chi2_thickness_var:.4f}")

# Calcular el valor p
# Para una prueba de cola izquierda, el valor p es la probabilidad de obtener un estadístico χ² tan extremo o más extremo (menor) que el calculado,
# asumiendo que H0 es verdadera.
p_value_thickness_var = chi2.cdf(chi2_statistic_thickness_var, degrees_of_freedom_thickness_var)

print(f"   Valor p: {p_value_thickness_var:.4f}")

# 6. Graficar la distribución Chi-cuadrado, región de rechazo y estadístico
x = np.linspace(0, chi2.ppf(0.999, degrees_of_freedom_thickness_var), 200) # Extender el rango para ver la cola derecha si es necesario
y = chi2.pdf(x, degrees_of_freedom_thickness_var)

fig, ax = plt.subplots()
ax.plot(x, y, 'r-', lw=2, label=f'Distribución χ² (gl={degrees_of_freedom_thickness_var})')

# Rellenar la región de rechazo (cola izquierda)
x_reject = np.linspace(0, critical_chi2_thickness_var, 100)
y_reject = chi2.pdf(x_reject, degrees_of_freedom_thickness_var)
ax.fill_between(x_reject, y_reject, color='blue', alpha=0.5, label=f'Región de Rechazo (α={alpha_thickness_var})')

# Dibujar el valor crítico
ax.axvline(critical_chi2_thickness_var, color='blue', linestyle='--', label=f'Valor Crítico χ² = {critical_chi2_thickness_var:.4f}')

# Dibujar el estadístico calculado
ax.axvline(chi2_statistic_thickness_var, color='green', linestyle='-', label=f'Estadístico χ² = {chi2_statistic_thickness_var:.4f}')

ax.set_xlabel('Valor χ²')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución χ² con Región de Rechazo y Estadístico de Prueba (Prueba de Cola Izquierda)')
ax.legend()
plt.grid(True)
plt.show()


# 7. Tomar una decisión estadística y concluir
print("\n4. Decisión Estadística:")
if chi2_statistic_thickness_var < critical_chi2_thickness_var:
    print(f"   Como el estadístico χ² ({chi2_statistic_thickness_var:.4f}) es menor que el valor crítico χ² ({critical_chi2_thickness_var:.4f}),")
    print("   se rechaza la hipótesis nula (H0).")
    decision_thickness_var = "Rechazar H0"
else:
    print(f"   Como el estadístico χ² ({chi2_statistic_thickness_var:.4f}) no es menor que el valor crítico χ² ({critical_chi2_thickness_var:.4f}),")
    print("   no se rechaza la hipótesis nula (H0).")
    decision_thickness_var = "No rechazar H0"

print("\nConclusión en términos prácticos:")
if decision_thickness_var == "Rechazar H0":
    print(f"Con un nivel de significancia de α = {alpha_thickness_var}, hay suficiente evidencia para concluir que la varianza poblacional en el grosor de las piezas metálicas es menor que {sigma2_population_thickness} mm².")
    print(f"El valor p ({p_value_thickness_var:.4f}) es menor que α ({alpha_thickness_var}), lo que apoya esta conclusión.")
else:
    print(f"Con un nivel de significancia de α = {alpha_thickness_var}, no hay suficiente evidencia para concluir que la varianza poblacional en el grosor de las piezas metálicas es menor que {sigma2_population_thickness} mm².")
    print(f"El valor p ({p_value_thickness_var:.4f}) es mayor que α ({alpha_thickness_var}), lo que apoya esta conclusión.")

print(f"\nSignificado del valor p:")
print(f"El valor p de {p_value_thickness_var:.4f} representa la probabilidad de observar una varianza muestral de {s2_sample_thickness_var} (o menor), asumiendo que la verdadera varianza poblacional es {sigma2_population_thickness}.")
print("Como el valor p es pequeño (menor que α), sugiere que nuestra observación es poco probable bajo la hipótesis nula, lo que nos lleva a rechazar H0." if decision_thickness_var == "Rechazar H0" else "Como el valor p es grande (mayor que α), sugiere que nuestra observación es probable bajo la hipótesis nula, lo que nos lleva a no rechazar H0.")
```
